```python
import cv2  # Import OpenCV library
import mediapipe as mp  # Import Mediapipe library
import time  # Import time module for time tracking

# Initialize mediapipe drawing utilities
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

# Open the video file
cap = cv2.VideoCapture('TEST.mp4')

# Initialize previous time variable
pTime = 0

# Loop through the video frames
while True:
    # Read the current frame from the video
    success, img = cap.read()

    # Convert the color space of the image from BGR to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Process the pose estimation on the RGB image
    results = pose.process(imgRGB)

    # Draw pose landmarks and connections if landmarks are detected
    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
        # Iterate through all detected landmarks
        for id, lm in enumerate(results.pose_landmarks.landmark):
            # Get the dimensions of the image
            h, w, c = img.shape
            # Print the landmark ID and its coordinates
            print(id, lm)
            # Calculate the landmark position in pixel coordinates
            cx, cy = int(lm.x * w), int(lm.y * h)
            # Draw a circle at the landmark position
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)

    # Calculate and display the frames per second (FPS) on the image
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)

    # Display the annotated image
    cv2.imshow("Image", img)

    # Wait for a key press and check if 'q' is pressed to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
